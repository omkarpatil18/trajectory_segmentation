```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
tasks = np.array(['put_item_in_drawer', 'put_shoes_in_box', 'ball_in_hoop', 'stack_blocks'])[[2,3,1,0]]
uni_res_250 = np.array([16, 2, 0.5, 4])[[2,3,1,0]]
uni_res_500 = np.array([16, 26, 78, 14])[[2,3,1,0]]
uni_res_750 = np.array([16, 74, 96, 32])[[2,3,1,0]]

stit_res_250 = np.array([28, 0.5, 6, 18])[[2,3,1,0]]
stit_res_500 = np.array([62, 74, 94, 34])[[2,3,1,0]]
stit_res_750 = np.array([68, 92, 90, 42])[[2,3,1,0]]

diff_250 = stit_res_250-uni_res_250
diff_500 = stit_res_500-uni_res_500
diff_750 = stit_res_750-uni_res_750

task_1 = [diff_250[0], diff_500[0], diff_750[0]]
task_2 = [diff_250[1], diff_500[1], diff_750[1]]
task_3 = [diff_250[2], diff_500[2], diff_750[2]]
task_4 = [diff_250[3], diff_500[3], diff_750[3]]
```

```python
print(tasks)
print(diff_250)
print(diff_500)
print(diff_750)
```

```python
plt.plot(task_1)
plt.plot(task_2)
plt.plot(task_3)
plt.plot(task_4)
```

```python
plt.rcParams.update({'font.size': 14})
fig, ax = plt.subplots(layout='constrained')
ax.bar([0, 0.2, 0.4, 0.6], diff_250, color=['orange', 'royalblue', 'palevioletred', 'forestgreen'], width=0.15, label=tasks) 
ax.bar([1, 1.2, 1.4, 1.6], diff_500, color=['orange', 'royalblue', 'palevioletred', 'forestgreen'], width=0.15) 
ax.bar([2, 2.2, 2.4, 2.6], diff_750, color=['orange', 'royalblue', 'palevioletred', 'forestgreen'], width=0.15)

ax.set_ylabel("Improvement in performance")
ax.set_xlabel("Number of datapoints")
plt.ylim(-10, 55)
ax.set_xticks(np.arange(len(["250", "500", "750"])) + 0.3, ["250", "500", "750"])
plt.legend(loc='best')
plt.savefig("/home/local/ASUAD/opatil3/res.png")
```
